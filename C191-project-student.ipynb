```python
#Task 2
#Get top 5 countries who has the least number of cases and plot a bar graph
#Covide Data 
import  numpy as np
import pandas as pd
from matplotlib import pyplot as plt


dataframe = pd.read_csv('covid19.csv')
df = pd.DataFrame(dataframe, columns = ['total_cases'])
df = dataframe.dropna()
df

sorted_df = df.sort_values(by = [ 'total_cases'])

lowest_cases = sorted_df['country'].tail(5)

print(sorted_df, lowest_cases)

name = lowest_cases['Official Name of Satellite']
weight = lowest_cases['Dry Mass (Kilograms)']

plt.xlabel('Official Name of Satellite')
plt.xticks(rotation = ("vertical"))
plt.ylabel('Dry Mass (Kilograms)')

label = name
value = weight
plt.bar(label, value,width=0.4, color=('red','yellow','green','pink','blue')) #bar-grap
```
